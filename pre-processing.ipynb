### Importing Libraries

````python
# importing libraries
from google.colab import drive
import os
import pandas as pd
from datetime import datetime
````

### Reading 132 transaction csv files from Google drive folder path: ```/content/drive/MyDrive/MLAA AT3/Dataset```

````python
# Mount your Google Drive
drive.mount('/content/drive')
````

````python
# Set the path to your CSV files folder
path = '/content/drive/MyDrive/MLAA AT3/Dataset'
````

````python
# Get a list of all CSV files in the folder
csv_files = [f for f in os.listdir(path) if f.endswith('.csv') and f != 'customers.csv']
````

````python
# Initialize an empty DataFrame `txn_data` to store the combined transaction data
txn_data = pd.DataFrame()
````

````python
# Loop through each CSV file and append its data to the combined data DataFrame
for file in csv_files:
    # Read the CSV file into a DataFrame
    df = pd.read_csv(os.path.join(path, file), delimiter='|')
    
    # Concatenate the DataFrame to the combined data DataFrame
    txn_data = pd.concat([txn_data, df], ignore_index=True)
````

````python
# Display the combined transaction data
txn_data
````

### Reading `customers.csv` file from Google drive folder

````python
# Reading customer file
customer_df = pd.read_csv(os.path.join(path, 'customers.csv'), delimiter='|')
````

### Feature engineering in Transaction dataset

````python
# unix_time to timestamp
txn_data['timestamp'] = pd.to_datetime(txn_data['unix_time'], unit='s')

# droping unix_time column
txn_data.drop('unix_time', axis=1, inplace=True)
````

### Feature engineering in Customer dataset

````python
# Convert the date of birth column to a datetime object in customer 
customer_df['dob'] = pd.to_datetime(customer_df['dob'])

# Calculate age
today = datetime.today()
customer_df['age'] = (today - customer_df['dob']).astype('<m8[Y]')

# drop unnecessary columns
customer_df.drop(['dob','first','last','ssn','street'], axis=1, inplace=True)
````

### Joining datasets

````python
# Merge the dataframes using two conditions using left join
final_df = pd.merge(txn_data, customer_df, how='left', on=['cc_num','acct_num'])
````
